### 1.1 Was Datenvisualisierung ist

Datenvisualisierung ist die graphische Darstellung von von Daten. Es ist ein wichtiges Werkzeug zur Datenanalyse, da es uns hilft Muster, Trends und Beziehungen in Datens√§tzen zu finden, die m√∂glicherweise nicht anhand von Tabellen ersichtlich sind. Unter anderem k√∂nnen wir durch Datenvisualisierung schnell Ausrei√üer identifizieren oder die Verteilung von Werten verstehen.

### 1.2 Warum Datenvisualisierung wichtig ist

Datenvisualisierung ist aus mehreren Gr√ºnden wichtig. Vor allem hilft sie uns die Daten zu verstehen, mit denen wir arbeiten. Durch Visualisierungen k√∂nnen wir schnell Trends und Muster erkennen, die wir durch das blo√üe Lesen von Tabellen nicht erkannt h√§tten.

Au√üerdem ist Datenvisualisierung wichtig wenn man Machine Learning Projekte starten m√∂chte. Bevor man ein Machine Learning Modell bildet, ist es wichtig den Datensatz und die zugrundeliegende Struktur zu verstehen. Visualisierung hilft uns dabei Daten in sinnvoll zu erkunden und so fundiert zu entscheiden, welche Merkmale des Datensatzes benutzt werden und wie der Datensatz vorverarbeitet werden sollte.

Dar√ºberhinaus ist Datenvisualisierung wichtig um Erkenntnisse mit anderen zu kommunizieren. Durch informative Visualisierungen k√∂nnen wir effektiv unsere Erkenntnisse kommunizieren und zum Beispiel unseren Vorgesetzen Empfehlungen geben, unabh√§ngig davon ob sie das technische Hintergrundwissen haben oder nicht.

## 2. Wann welche Visualisierung verwendet wird ü§î

Es gibt verschiedene Arten der Datenvisualisierung. Die drei g√§ngigsten Arten sind *Balken-, Linien- , und  Streudiagramme*. Wann welche Art typischerweise verwendet wird, wird im folgendem erkl√§rt.

1.  **Balkendiagramme (bar plot)**: Balkendiagramme sind f√ºr den Vergleich einzelner Kategorien oder Gruppen geeignet. Sie sind besonders n√ºtzlich, wenn man den Unterschied zwischen zwei oder mehreren *Gruppen* visualisieren m√∂chten. Du k√∂nntest beispielsweise ein Balkendiagramm verwenden, um die Verkaufszahlen f√ºr verschiedene Produkte oder die Anzahl der Auszubildenden, die eine Pr√ºfung bestanden oder nicht bestanden haben, zu vergleichen.
    
2.  **Liniendiagramme (line plot)**: Liniendiagramme eignen sich zur Visualisierung von Trends im Zeitverlauf oder √ºber verschiedene Stufen einer Variablen. Sie sind besonders n√ºtzlich, wenn man sehen m√∂chten, wie sich eine Variable im Laufe der Zeit oder √ºber verschiedene Kategorien hinweg ver√§ndert. Du k√∂nntesn beispielsweise ein Liniendiagramm verwenden, um zu zeigen, wie sich die Einnahmen deines Unternehmens in den letzten Jahren ver√§ndert haben.
    
3.  **Streudiagramme (scatter plot)**:  Streudiagramme sind n√ºtzlich, um die Beziehung zwischen zwei *kontinuierlichen Variablen* zu visualisieren. Sie sind besonders n√ºtzlich, wenn man sehen m√∂chten, wie zwei Variablen miteinander in Beziehung stehen oder ob es eine Korrelation zwischen ihnen gibt. Du k√∂nntest beispielsweise ein Streudiagramm verwenden, um festzustellen, ob es eine Beziehung zwischen der Gr√∂√üe und dem Gewicht einer Person gibt.
    
Das hier sind nur allgemeine Richtlinien und es kann Ausnahmen geben. Es ist immer eine gute Idee, die Art Deiner Daten und das, was Du kommunizieren willst, zu ber√ºcksichtigen, bevor Du dich f√ºr Art der Darstellung entscheidest.

### 4.3 Liniendiagramm | Durschnittstemperatur in Deutschland

Wir zeigen euch abschlie√üen noch schnell, wie ihr mit Matplotlib und Seaborn ein Liniendiagramm erstellt, das die durchschnittliche Temperatur in Deutschland darstellt.

#### 4.3.1 Liniendiagramm mit Matplotlib

````python
import matplotlib.pyplot as plt
%matplotlib inline

# x- und y-Achse f√ºr das Diagramm bestimmen
x = temp_df['decade']
y = temp_df['AverageTemperature']

#  Liniendiagramm erstellen
plt.plot(x, y)

# Titel und Achsenbeschriftung bestimmen
plt.title('Durschnittstemperatur in Deutschland (1750-2020)')
plt.xlabel('Jahrzent')
plt.ylabel('Temperatur (¬∞C)')

# Diagramm anzeigen lassen
plt.show()
````

#### 4.3.2 Liniendiagramm mit Seaborn

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Liniendiagramm erstellen
sns.lineplot(x="decade", y="AverageTemperature", data=temp_df)

# Achsensbeschriftung hinzuf√ºgen
plt.title('Durschnittstemperatur in Deutschland (1750-2020)')
plt.xlabel("Jahr")
plt.ylabel("Temperatur (Celsius)")

# Diagramm anzeigen lassen
plt.show()
````
